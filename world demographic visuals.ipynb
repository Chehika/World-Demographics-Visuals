```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/world_demographics.csv")

aland_2018 = df[(df["Country or Area"] == "Åland Islands") & (df["Year"] == 2018)].copy()

# converting age and value to numeric
aland_2018.loc[:, "Age"] = pd.to_numeric(aland_2018["Age"], errors='coerce')
aland_2018.loc[:, "Value"] = pd.to_numeric(aland_2018["Value"], errors='coerce')

# Histogram: Age Distribution (Both Sexes)
age_data = aland_2018[(aland_2018["Sex"] == "Both Sexes") & (aland_2018["Age"].notnull())]

plt.figure(figsize=(8, 5))
plt.bar(age_data["Age"], age_data["Value"], width=1.0, color="lightgreen", edgecolor="black")
plt.title("Age Distribution in Åland Islands (2018)")
plt.xlabel("Age")
plt.ylabel("Population")
plt.tight_layout()
plt.show()
```
